```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train_merge.csv")

test = pd.read_csv("../testing_phase1/volume(table 6)_test2.csv")
```

```python
all_data = pd.concat((train.copy(), test.copy()))
all_data["time"] = all_data["time"].apply(lambda x: pd.Timestamp(x))
all_data["count"] = 1
all_data.index = all_data.time
all_data0 = all_data[(all_data["tollgate_id"] == 3) & (all_data["direction"] == 0)][["time","count"]]
del all_data0["time"]
```

```python
all_data0 = all_data0.resample("20T").sum()
all_data0["time"] = all_data0.index
all_data0["hour"] = all_data0["time"].apply(lambda x: x.hour)
all_data0["minute"] = all_data0["time"].apply(lambda x: x.minute)
all_data0["day"] = all_data0["time"].apply(lambda x: x.day)
all_data0["month"] = all_data0["time"].apply(lambda x: x.month)
del all_data0["time"]
all_data0
```
