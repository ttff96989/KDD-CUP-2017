```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test_df = pd.read_csv("./testing_phase1/volume(table 6)_test2.csv")
test_df.head()
```

```python
# 101030201030103010313010281030
#               3110281030
test1 = test_df.copy()
tollgate_id = 2
direction = 0
test1["time"] = test1["time"].apply(pd.Timestamp)
test1.index = test1["time"]
test1["volume"] = 1
test1["day"] = test1["time"].apply(lambda x: x.day)
for offset in range(7):
    temp = test1[(test1["tollgate_id"] == tollgate_id) & 
                (test1["direction"] == direction) & 
                (test1["day"] == 25 + offset)][["volume"]].resample("20T").sum().dropna()
    temp.plot()
```
